```python
import pandas as pd
import numpy as np
```

```python
messages = pd.read_csv("SpamClassifier-master/SMSSpamCollection.txt",sep="\t",names=["label","massage"])
```

```python
messages.head()
```

```python
import re
from nltk.corpus import stopwords
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []

for i in range(0,len(messages)):
    review = re.sub("^a-zA-Z"," ",messages["massage"][i])  # here if the word does not contain alphabets from a to z OR A to Z just remove that word (Special charecter detection)
    review = review.lower()
    review = review.split()
    review =[ps.stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

# Create Bag Of Words

```python
## Creating the Bag of word model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv = CountVectorizer(max_features=500,binary=True)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
print(X.shape)
```

```python
cv.vocabulary_ ## here we can get the top 500 Most reapeted words from Dataset
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))
X
```

# Using the N-Grams Parameter

```python
## Creating the Bag of word model
from sklearn.feature_extraction.text import CountVectorizer
## for Binary BOW enable binary=True
cv = CountVectorizer(max_features=500,binary=True,ngram_range=(1,2))
X = cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
X
```

# Create TF-IDF And NGrams

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=100)
X=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000,
    formatter=dict(float=lambda x: "%.3g" % x))
```
